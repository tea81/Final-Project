Based of the above data we can see that numeric values are stored as objects. This required us to dive further into the data. Initially, we found that some key columns included percent signs, so we removed those. Additionally, we had to address non-numeric values in the columns to avoid errors during conversion to numeric types.

```python
#replace all other strings  without % in target column with nan before removing the percents
col = 'Percentage of Grade 3 Students Achieving the Provincial Standard in Reading'
count_percent = third_grade_scores[col].str.contains('%').sum()

count_per = 0
count_non_per = 0

for i, value in third_grade_scores[col].items():
    if '%' not in str(value):  # Convert the value to string and check if it contains '%'
        third_grade_scores.at[i, col] = np.nan
        count_non_per += 1
    else:
        count_per += 1


print("Number of Values with Percents: ", count_per)
print("Number of Values without Percents: ", count_non_per)
```

```python
#drop the null values
third_grade_scores.dropna(inplace = True)
third_grade_scores.isnull().sum()
```

```python
#Remove percent signs in target column
third_grade_scores[col] = third_grade_scores[col].str.replace('%', '').astype(float)
third_grade_scores.dtypes
```

Next we converted the other non-numeric columns values to NaN.

```python
expected_num_col = [
    'Enrolment',
    'Percentage of Students Whose First Language Is Not English',
    'Percentage of Students Whose First Language Is Not French',
    'Percentage of Students Who Are New to Canada from a Non-English Speaking Country',
    'Percentage of Students Who Are New to Canada from a Non-French Speaking Country',
    'Percentage of Students Receiving Special Education Services',
    'Percentage of Students Identified as Gifted',
    'Percentage of Grade 3 Students Achieving the Provincial Standard in Reading',
    'Percentage of School-Aged Children Who Live in Low-Income Households',
    'Percentage of Students Whose Parents Have No Degree, Diploma or Certificate'
]
#'Percentage of Students Receiving Special Education Services'--------This column is a tupple for some reason

for col in expected_num_col:
    third_grade_scores[col] = pd.to_numeric(third_grade_scores[col],errors='coerce')
third_grade_scores.dtypes
```

```python
#Determine how many null values there are
third_grade_scores.isnull().sum()
```

```python
#Drop the null values since there is a small amount
third_grade_scores.dropna(inplace = True)
third_grade_scores.isnull().sum()
```

```python
#Understand the dataset to see the correlation between each variable
numeric_col = [
    'Enrolment',
    'Percentage of Students Whose First Language Is Not English',
    'Percentage of Students Whose First Language Is Not French',
    'Percentage of Students Who Are New to Canada from a Non-English Speaking Country',
    'Percentage of Students Who Are New to Canada from a Non-French Speaking Country',
    'Percentage of Students Receiving Special Education Services',
    'Percentage of Students Identified as Gifted',
    'Percentage of Grade 3 Students Achieving the Provincial Standard in Reading',
    'Percentage of School-Aged Children Who Live in Low-Income Households',
    'Percentage of Students Whose Parents Have No Degree, Diploma or Certificate'
]
a = third_grade_scores[numeric_col].corr()
plt.rcParams['figure.figsize']=(12,8)
ax = sns.heatmap(a, linewidth=0.5, cmap= 'BuGn_r', annot = True)
plt.show()
```

Based off the correlation map, we chose to explore relations between the variables that had a strong postive or negative correlation.

The heatmap indicates that the columns Percentage of School-Aged Children Who Live in Low-Income Households and Percentage of Students Whose Parents Have No Degree, Diploma, or Certificate, show a moderate positive correlation (0.53). This indicates that there might be a relationship between lower parental education levels and living in low-income households. We used a scatter plot to visualize the relationship.

```python
sns.set(style="whitegrid")
plt.figure(figsize=(25, 7))
#Visualize correlations
ax = sns.scatterplot(
    x=third_grade_scores['Percentage of School-Aged Children Who Live in Low-Income Households'],
    y=third_grade_scores['Percentage of Students Whose Parents Have No Degree, Diploma or Certificate'],
    data=third_grade_scores,
    color="darkgreen",
    s=100,
    alpha=0.7
)

# Add a title and customize font sizes
plt.title('Relationship Between Low-Income Households and Parental Education', fontsize=20)
plt.xlabel('Percentage of School-Aged Children Who Live in Low-Income Households', fontsize=14)
plt.ylabel('Percentage of Parents Have No Higher Education', fontsize=14)
plt.show()
```

Percentage of Grade 3 Students Achieving the Provincial Standard in Reading has a negative correlation with Percentage of School-Aged Children Who Live in Low-Income Households (-0.34). This suggests that higher percentages of children living in low-income households might be associated with lower achievement in reading.
